```python
from __future__ import print_function
import pandas as pd
import shutil
import os
import sys

labels = pd.read_csv(r'C:\Users\styli\Downloads\RGB_Augmented\mm.csv')

train_dir =r'C:\Users\styli\Downloads\RGB_Augmented\Train_Images'
DR = r"C:\Users\styli\Downloads\RGB_Augmented\DR"
if not os.path.exists(DR):
    os.mkdir(DR)
#name=[]
#for x,y in labels.values:
#    name.append(x)
#    break
#print(name[0].split('/')[1])
for filename, class_name in labels.values:
    # Create subdirectory with `class_name`
    if not os.path.exists(DR + str(class_name)):
        os.mkdir(DR + str(class_name))
    filename=filename.split('/')[1]
    src_path = train_dir + '/'+ filename
    dst_path = DR + str(class_name) + '/' + filename
    try:
        shutil.copy(src_path, dst_path)
        print("sucessful")
    except IOError as e:
        print(e)
    except:
        print('When try copy file {} to {}, unexpected error: {}'
              .format(src_path, dst_path, sys.exc_info()))
```
